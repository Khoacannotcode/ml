## 1. Mô hình toán học của Supervised Learning

### Định nghĩa

**Supervised learning** là thuật toán dự đoán đầu ra (outcome) của một dữ liệu mới (new input) dựa trên các cặp (input, outcome) đã biết từ trước. Cặp dữ liệu này còn được gọi là (data, label), tức (dữ liệu, nhãn). Supervised learning là nhóm phổ biến nhất trong các thuật toán Machine Learning.

Một cách toán học, Supervised learning là khi chúng ra có một tập hợp biến đầu vào $X=\{{x_1,x_2,...,x_N}\}$ và một hợp nhãn tương ứng $X=\{{y_1,y_2,...,y_N}\}$, trong đó $(x_i,y_i)$ là các vector. Các cặp dữ liệu biết trước $(x_i, y_i)\in X\times Y$ được gọi là tập training data(dữ liệu huấn luyện). Từ tập training data này, chúng ta cần tạo ra hàm số ánh xạ mỗi phần tử từ tập $X$ sang một phần tử(xấp xỉ) tương ứng của tập $Y$.
#### $$y_i \approx f(x_i), \forall i = 1,2,...,N$$

Mục đích là xấp xỉ hàm số f thật tốt để khi có một dữ liệu x mới, chúng ta có thể tính được nhãn tương ứng của nó $y=f(x)$.

### *Ví dụ: Bài toán nhận dạng chữ viết tay*

Trong nhận dạng chữ viết tay, ta có ảnh của hàng nghìn ví dụ của mỗi chữ số được viết bởi nhiều người khác nhau. ### Simple Linear Regression

Đây là một trong những loại kỹ thuật hồi quy phổ biến và thú vị nhất. Ở đây chúng ta **dự đoán một biến mục tiêu Y dựa trên biến đầu vào X**. Một **mối quan hệ tuyến tính nên tồn tại giữa biến mục tiêu và yếu tố dự đoán** và do đó có tên là Hồi quy tuyến tính.
Xem xét dự đoán mức lương của một nhân viên dựa trên tuổi của anh ấy/cô ấy. Chúng ta có thể dễ dàng xác định rằng dường như có một mối tương quan giữa tuổi của nhân viên và mức lương (nhiều tuổi hơn là tiền lương). Giả thuyết về hồi quy tuyến tính là:
### $$Y = a+bX$$

trong đó, Y đại diện cho tiền lương, X là tuổi nhân viên và a, b là các hệ số của phương trình. Vì vậy, để dự đoán Y (tiền lương) cho X (tuổi), chúng ta cần biết các giá trị của a và b (các hệ số mô hình).

![Example of Linear Regression](https://miro.medium.com/max/445/1*Bkh-4Dwu0nQtPGwqFXANtQ.png)

Trong khi training và build mô hình hồi quy(regression model), chính các hệ số này được học và trang bị cho dữ liệu training. Mục đích của quá trình training là **tìm ra một đường thẳng phù hợp nhất sao cho hàm chi phí tối tiểu**. Hàm chi phí chính là sai số so với giá trị thực tế. Trong quá trình đào tạo, chúng ta cố gắng giảm thiểu error giữa các giá trị thực tế và dự đoán và do đó tối thiểu hàm chi phí(cost function).

Trong hình trên, các điểm màu đỏ là các điểm dữ liệu và đường màu xanh là đường dự đoán cho dữ liệu huấn luyện. Để có được giá trị dự đoán, các điểm dữ liệu này được chiếu lên đường thẳng.

Tóm lại, mục tiêu của chúng ta là tìm ra các giá trị hệ số như vậy sẽ giảm thiểu hàm chi phí. Hàm chi phí phổ biến nhất là **`Mean Square Error(MSE)`** bằng với độ lệch bình phương trung bình giữa một giá trị thực tế quan sát và dự đoán. Các giá trị hệ số có thể được tính bằng cách sử dụng phương pháp **Gradient Descent** sẽ được thảo luận chi tiết trong chương sau. Để hiểu ngắn gọn, Gradient, chúng ta bắt đầu với một số giá trị ngẫu nhiên của các hệ số, tính toán độ dốc của hàm chi phí trên các giá trị này, cập nhật các hệ số và tính lại hàm chi phí. Quá trình này được lặp lại cho đến khi chúng ta tìm thấy một giá trị làm cho hàm chi phí(hàm mất mát) đạt nhỏ nhất toàn cục.

### Polynomial Regression

Trong Polynimial (hồi quy đa thức), chúng ta chuyển đổi các đặc trưng ban đầu thành các tính năng trong đa thức của một mức độ nhất định và sau đó áp dụng hồi quy tuyến tính trên nó. Hãy xem xét mô hình tuyến tính ở trên Y = a + bX được chuyển đổi như sau:
### $$Y = a + bX + cX^2$$

**Nó vẫn là một mô hình tuyến tính nhưng đường cong bây giờ là bậc hai chứ không phải là một đường. Scikit-Learn cung cấp lớp PolynomialFeatures để biến đổi các đặc trưng.**

![Polynomial Regression Graph Example](https://miro.medium.com/max/1181/1*09J83FIdzj2VM8Hy88OP0A.png)

Nếu chúng ta tăng bâc lên một giá trị rất cao, đường cong sẽ Overfiting vì nó học luôn những thứ với vẫn trong dữ liệu trong dữ liệu. Vấn đề Overfiting sẽ được bàn trong các chương sau.

### Support Vector Regression

Trong SVR, chúng ta xác định một siêu phẳng(hyperplane) có lề tối đa sao cho số điểm dữ liệu tối đa nằm trong lề đó. Nó tương tự như SVM classification.

Thay vì giảm thiểu tỷ lệ lỗi như trong hồi quy tuyến tính, chúng ta cố gắng điều chỉnh lỗi trong một ngưỡng nhất định. Mục tiêu của chúng tôi trong SVR về cơ bản là xem xét các điểm nằm trong lề. Line phù hợp nhất của chúng ta là siêu phẳng có số điểm tối đa.

### Decision Tree Regression

Decision Tree có thể được sử dụng để phân loại cũng như hồi quy. Trong Decision Trees, ở mỗi cấp độ, chúng ta cần xác định thuộc tính tách. Trong trường hợp hồi quy, thuật toán ID3 có thể được sử dụng để xác định nút tách bằng cách giảm độ lệch chuẩn (trong thông tin phân loại được sử dụng).

Cây quyết định được xây dựng bằng cách phân vùng dữ liệu thành các tập hợp con chứa các thể hiện có giá trị tương tự (đồng nhất). Độ lệch chuẩn được sử dụng để tính độ đồng nhất của mẫu số. Nếu mẫu số hoàn toàn đồng nhất, độ lệch chuẩn của nó bằng không.

![Decision Tree Regression](https://scikit-learn.org/stable/_images/sphx_glr_plot_tree_regression_001.png)

Các bước để tìm nút chia được mô tả ngắn gọn như dưới đây:

1. Tính độ lệch chuẩn của biến mục tiêu:
#### $$Stardarddeviation= S = \sqrt{\left(\frac{\sum (x-\bar x)^2}{n}\right)}$$

2. Tách tập dữ liệu trên các thuộc tính khác nhau và tính độ lệch chuẩn cho từng nhánh (độ lệch chuẩn cho mục tiêu và dự đoán). Giá trị này được trừ khỏi độ lệch chuẩn trước khi chia. Kết quả là sự giảm độ lệch chuẩn.
#### $$SDR(T,X)=S(T)-S(T,X)$$

3. Thuộc tính có mức giảm độ lệch chuẩn lớn nhất được chọn làm nút chia.

4. Tập dữ liệu được chia dựa trên các giá trị của thuộc tính được chọn. Quá trình này được chạy đệ quy trên các nhánh không có lá, cho đến khi tất cả dữ liệu được xử lý.

**Để tránh Overfiting, Coefficient of Deviation (CV) được sử dụng để quyết định khi nào dừng phân nhánh**. *Cuối cùng, trung bình của mỗi nhánh được gán cho nút lá liên quan (theo phương pháp hồi quy có nghĩa là được thực hiện trong trường hợp như trong chế độ phân loại của các nút lá được thực hiện).*

### Random Forest Regression

Random Forest Regression là một cách tiếp cận đồng bộ, trong đó chúng ta tính đến các dự đoán của một số Decision Regression Tree.

1. **Select K random points.**
2. **Identify n where n is the number of decision tree regressors to be created. Repeat step 1 and 2 to create several regression trees.**
3. **The average of each branch is assigned to leaf node in each decision tree.**
4. **To predict output for a variable, the average of all the predictions of all decision trees are taken into consideration.**

*Random forest Regression ngăn chặn Overfiting (thường thấy trong các Decision Tree) bằng cách tạo các tập hợp con ngẫu nhiên của các đặc trưng và xây dựng các cây nhỏ hơn bằng cách sử dụng các tập hợp con này.*
